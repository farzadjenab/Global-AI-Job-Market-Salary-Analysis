```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the datasets
df1 = pd.read_csv("Global AI Job Market & Salary (1).csv")
df2 = pd.read_csv("Global AI Job Market & Salary (2).csv")

# Show the first few rows of each dataset
df1.head(), df2.head()
```

```python
# Checking the structure of the datasets
df1_info = df1.info()
df2_info = df2.info()

# Checking for missing values
df1_missing = df1.isnull().sum()
df2_missing = df2.isnull().sum()

# Show basic statistics (numerical columns)
df1_stats = df1.describe()
df2_stats = df2.describe()

df1_info, df2_info, df1_missing, df2_missing, df1_stats, df2_stats
```

```python
# Let's check the distribution of salaries
plt.figure(figsize=(12, 6))
plt.hist(df1['salary_usd'], bins=30, alpha=0.5, label='Dataset 1')
plt.hist(df2['salary_usd'], bins=30, alpha=0.5, label='Dataset 2')
plt.title("Salary Distribution for both datasets")
plt.xlabel("Salary (USD)")
plt.ylabel("Frequency")
plt.legend(loc='upper right')
plt.show()

# Checking the experience level distribution
experience_level_1 = df1['experience_level'].value_counts()
experience_level_2 = df2['experience_level'].value_counts()

experience_level_1, experience_level_2
```

```python
# Analyzing company location and employee residence
company_location_1 = df1['company_location'].value_counts()
employee_residence_1 = df1['employee_residence'].value_counts()

company_location_2 = df2['company_location'].value_counts()
employee_residence_2 = df2['employee_residence'].value_counts()

# Displaying the distributions
company_location_1, employee_residence_1, company_location_2, employee_residence_2
```

```python
# Checking how remote work ratio is distributed
remote_ratio_1 = df1['remote_ratio'].value_counts()
remote_ratio_2 = df2['remote_ratio'].value_counts()

remote_ratio_1, remote_ratio_2
```

```python
# Checking the most common required skills
skills_1 = df1['required_skills'].str.split(',').explode().value_counts().head(10)
skills_2 = df2['required_skills'].str.split(',').explode().value_counts().head(10)

# Displaying the top 10 most common skills for both datasets
skills_1, skills_2
```

```python
# Convert posting_date and application_deadline to datetime format
df1['posting_date'] = pd.to_datetime(df1['posting_date'])
df1['application_deadline'] = pd.to_datetime(df1['application_deadline'])

df2['posting_date'] = pd.to_datetime(df2['posting_date'])
df2['application_deadline'] = pd.to_datetime(df2['application_deadline'])

# Filter jobs with application deadline passed
expired_jobs_1 = df1[df1['application_deadline'] < pd.Timestamp.today()]
expired_jobs_2 = df2[df2['application_deadline'] < pd.Timestamp.today()]

# Displaying the number of expired jobs in each dataset
len(expired_jobs_1), len(expired_jobs_2)
```

```python
# Analyzing company size distribution
company_size_1 = df1['company_size'].value_counts()
company_size_2 = df2['company_size'].value_counts()

company_size_1, company_size_2
```

```python
# Checking the distribution of benefits scores
plt.figure(figsize=(12, 6))
plt.hist(df1['benefits_score'], bins=20, alpha=0.5, label='Dataset 1')
plt.hist(df2['benefits_score'], bins=20, alpha=0.5, label='Dataset 2')
plt.title("Benefits Score Distribution")
plt.xlabel("Benefits Score")
plt.ylabel("Frequency")
plt.legend(loc='upper right')
plt.show()

# Checking the mean benefits score for both datasets
mean_benefits_1 = df1['benefits_score'].mean()
mean_benefits_2 = df2['benefits_score'].mean()

mean_benefits_1, mean_benefits_2
```

```python
# Grouping by job title and calculating average salary
avg_salary_by_job_1 = df1.groupby('job_title')['salary_usd'].mean().sort_values(ascending=False).head(10)
avg_salary_by_job_2 = df2.groupby('job_title')['salary_usd'].mean().sort_values(ascending=False).head(10)

avg_salary_by_job_1, avg_salary_by_job_2
```

```python
# Printing out final summary
summary_1 = df1.describe()
summary_2 = df2.describe()

summary_1, summary_2
```

```python
# Grouping by job title and analyzing salary distribution for both datasets
salary_by_job_1 = df1.groupby('job_title')['salary_usd'].describe()
salary_by_job_2 = df2.groupby('job_title')['salary_usd'].describe()

# Displaying the salary summary for different job titles
salary_by_job_1, salary_by_job_2
```

```python
# Scatter plot to visualize the relationship between years of experience and salary
plt.figure(figsize=(12, 6))

# Dataset 1: Experience vs Salary
plt.subplot(1, 2, 1)
plt.scatter(df1['years_experience'], df1['salary_usd'], alpha=0.5)
plt.title("Experience vs Salary (Dataset 1)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (USD)")

# Dataset 2: Experience vs Salary
plt.subplot(1, 2, 2)
plt.scatter(df2['years_experience'], df2['salary_usd'], alpha=0.5)
plt.title("Experience vs Salary (Dataset 2)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary (USD)")

plt.tight_layout()
plt.show()

# Calculate correlation between years of experience and salary
corr_1 = df1['years_experience'].corr(df1['salary_usd'])
corr_2 = df2['years_experience'].corr(df2['salary_usd'])

corr_1, corr_2
```

```python
# Grouping by company location and calculating average salary
salary_by_location_1 = df1.groupby('company_location')['salary_usd'].mean().sort_values(ascending=False)
salary_by_location_2 = df2.groupby('company_location')['salary_usd'].mean().sort_values(ascending=False)

# Displaying the average salary by location for both datasets
salary_by_location_1, salary_by_location_2
```

```python
# Checking the difference between posting date and application deadline
df1['posting_date'] = pd.to_datetime(df1['posting_date'])
df1['application_deadline'] = pd.to_datetime(df1['application_deadline'])
df1['days_to_deadline'] = (df1['application_deadline'] - df1['posting_date']).dt.days

df2['posting_date'] = pd.to_datetime(df2['posting_date'])
df2['application_deadline'] = pd.to_datetime(df2['application_deadline'])
df2['days_to_deadline'] = (df2['application_deadline'] - df2['posting_date']).dt.days

# Checking the average time to deadline
avg_days_to_deadline_1 = df1['days_to_deadline'].mean()
avg_days_to_deadline_2 = df2['days_to_deadline'].mean()

avg_days_to_deadline_1, avg_days_to_deadline_2
```

```python
# Analyzing benefits_score by job title
benefits_by_job_1 = df1.groupby('job_title')['benefits_score'].mean().sort_values(ascending=False)
benefits_by_job_2 = df2.groupby('job_title')['benefits_score'].mean().sort_values(ascending=False)

# Displaying the benefits score for each job title
benefits_by_job_1, benefits_by_job_2
```

```python
# Grouping by education_required and calculating average salary
salary_by_education_1 = df1.groupby('education_required')['salary_usd'].mean().sort_values(ascending=False)
salary_by_education_2 = df2.groupby('education_required')['salary_usd'].mean().sort_values(ascending=False)

# Displaying the average salary by education required for both datasets
salary_by_education_1, salary_by_education_2
```

```python
# Grouping by employment type and calculating average salary
salary_by_employment_type_1 = df1.groupby('employment_type')['salary_usd'].mean().sort_values(ascending=False)
salary_by_employment_type_2 = df2.groupby('employment_type')['salary_usd'].mean().sort_values(ascending=False)

# Displaying the average salary by employment type for both datasets
salary_by_employment_type_1, salary_by_employment_type_2
```

```python
# Analyzing the most common required skills
skills_1 = df1['required_skills'].str.split(',').explode().value_counts().head(10)
skills_2 = df2['required_skills'].str.split(',').explode().value_counts().head(10)

# Displaying the top 10 most common skills for both datasets
skills_1, skills_2
```

```python
# Merge the two datasets
merged_df = pd.concat([df1, df2], ignore_index=True)

# Compare overall job statistics by job title
job_comparison = merged_df.groupby('job_title').agg({
    'salary_usd': 'mean',
    'years_experience': 'mean',
    'benefits_score': 'mean',
    'remote_ratio': 'mean'
}).sort_values(by='salary_usd', ascending=False)

# Displaying the comparison of jobs
job_comparison.head(10)
```

```python
# Display final summary of important insights
summary = {
    "Average Salary (Dataset 1)": df1['salary_usd'].mean(),
    "Average Salary (Dataset 2)": df2['salary_usd'].mean(),
    "Average Benefits Score (Dataset 1)": df1['benefits_score'].mean(),
    "Average Benefits Score (Dataset 2)": df2['benefits_score'].mean(),
    "Top Job Title in Dataset 1": df1.groupby('job_title')['salary_usd'].mean().idxmax(),
    "Top Job Title in Dataset 2": df2.groupby('job_title')['salary_usd'].mean().idxmax(),
}

summary
```
